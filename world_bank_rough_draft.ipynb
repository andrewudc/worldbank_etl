# Indicators

See: https://data.worldbank.org/indicator for indicators

If you click on a specific indicator, you'll find the indicator code in the web address. You can then use this code in the API like so:

http://api.worldbank.org/v2/countries/br/indicators/GC.DOD.TOTL.GD.ZS/?date=2006:2008&format=json

Where br is the country code for Brazil, GC.DOD.TOTL.GD.ZS is the indicator code, and date is the range of the year of interest.

Here are some possible indicators of interest although the API has many more:

| indicator | description |

# Get Indicator Data

```python
def indicator_data(indicator, country_code):
    url_19582008 = 'http://api.worldbank.org/v2/countries/' + country_code + '/indicators/' + indicator + '/?date=1968:2018&format=json'
    r = requests.get(url_19582008)
        
    json_results = r.json()[1]
    
    indicator_dict = defaultdict(int)
    
    for result in json_results:
        indicator_dict[result['date']] = result['value']
        
    return indicator_dict
```

# Make a Dictionary of Country ISO symbols

```python
def get_countries(page):
    country_url = 'http://api.worldbank.org/v2/countries?format=json'
    country_url = country_url + '&page=' + str(page)
    return requests.get(country_url)

def countryiso():
    countries = get_countries(1)
    number_pages = countries.json()[0]['pages']

    iso = defaultdict(str)
    nocapitol = []
    
    for i in range(1, number_pages + 1):
        country_data = get_countries(i)
        for country in country_data.json()[1]:
            if country['capitalCity'] != '':
                name = country['name']
                code = country['iso2Code']
                temp_dict = defaultdict(str)
                iso[name] = {'iso':code}
            else: # if entry does not have capital city (check for bad data)
                nocapitol.append(country['name'])    
        time.sleep(2)
    
    return iso, nocapitol

iso, nocapitol = countryiso()
```

# Scrape Project Country Name Data

```python
def scrapecountryname():
    
    html_doc = 'http://search.worldbank.org/api/v2/projects'
        
    with urllib.request.urlopen(html_doc) as url:
        r = url.read()
    
    soup = BeautifulSoup(r, 'html.parser')

    country_list_projects_api = []
    
    for country in soup.find("select", id="countryshortname").find_all("option"):
        country_list_projects_api.append(country['value'])
    
    return country_list_projects_api

country_list_projects_api = scrapecountryname()
```

# Match Project API Country Name with General API Country Name

```python
def match_project_country(iso):
    countries_not_in_project_api = []

    for country in iso.keys():
        proj_api_country = country.replace(' ','+')
        proj_api_country = proj_api_country.replace(',', '%2C')
        proj_api_country = proj_api_country.replace("'", '%27')
        proj_api_country = proj_api_country.replace("Dem.", 'Democratic')
        proj_api_country = proj_api_country.replace("Rep.", 'Republic+of')
        proj_api_country = proj_api_country.replace("Fed.", 'Federated')
        proj_api_country = proj_api_country.replace("Sts.", 'States+of')
        proj_api_country = proj_api_country.replace("Lao+PDR", "Lao+People%27s+Democratic+Republic")
        proj_api_country = proj_api_country.replace("Venezuela%2C+RB", "Venezuela%2C+Republica+Bolivariana+de")
        if proj_api_country not in country_list_projects_api:
            countries_not_in_project_api.append(country)
        elif proj_api_country in country_list_projects_api:
            iso[country]['project_api_name'] = proj_api_country
            
    return iso, countries_not_in_project_api

iso, no_match = match_project_country(iso)
```

```python
def create_url(rows, os, country_name):
    return 'http://search.worldbank.org/api/v2/projects?format=json&fl=id,countryname,prodline,status,project_name,boardapprovaldate,closingdate,totalamt,countrycode,countryshortname,keywords,project_abstract,majorsector_percent,sector,sector1,sector2,sector3,sector4,sector5,mjsector1,mjsector2,mjsector3,mjsector4,mjsector5&countryshortname_exact='+str(country_name)+'&source=IBRD&os='+str(os)+'&rows='+str(rows)

def project_country(country_name, iso_symbol, indicator1, indicator2, db_file):
    
    # Connect to SQLite database
    conn = sqlite3.connect(db_file)
    c = conn.cursor()
    id_column = 'projectid'
    table_name = 'project_data'

    # Start Getting Data from the API
    url = create_url(1, 0, country_name)
    print(url)
    r = requests.get(create_url(1, 0, country_name))
    
    time.sleep(3)

    total = int(r.json()['total'])
    rows = 500
    os = 0

    indicator_one = indicator_data(indicator1, iso[country_name]['iso'])
    time.sleep(2)
    indicator_two = indicator_data(indicator2, iso[country_name]['iso'])
    
    results = []
    
    while (os < total):
        r = requests.get(create_url(rows, os, country_name))
        print(os)

        os = os + rows
        
        time.sleep(5)

        current_data = r.json()['projects']
        
        for project in current_data:

            project = current_data[project]

            keys_of_interest = ['id', 'project_name', 'countrycode', 'countryshortname', 'totalamt', 'boardapprovaldate', 
                                'sector1', 'sector2', 'sector3', 'sector4', 'sector5']

            data_line = []
            
            for key in keys_of_interest:
                if key in project:
                    if key.find('sector') != -1:
                        data_line.append(project[key]['Name'])
                        data_line.append(project[key]['Percent'])
                    else:
                        data_line.append(project[key])
                else:
                    if key.find('sector') != -1:
                        data_line.append('None')
                        data_line.append('None')
                    else:
                        data_line.append('None')
            
            if 'boardapprovaldate' in project:
                year = project['boardapprovaldate'][0:4]
                if indicator_one[year]:
                    data_line.append(indicator_one[year])
                else:
                    data_line.append('None')
                if indicator_two[year]:
                    data_line.append(indicator_two[year])
                else:
                    data_line.append('None')
            else:
                data_line.append('None')
                data_line.append('None')
            
            results.append(data_line)
            print(tuple(data_line))
        
            try:
                c.execute('INSERT INTO ' + table_name + ' VALUES ' + str(tuple(data_line)))
            except sqlite3.IntegrityError:
                print('ERROR: ID already exists in PRIMARY KEY column {}'.format(id_column))

        
        time.sleep(3)
    
    conn.commit()
    conn.close()
    return results
```
